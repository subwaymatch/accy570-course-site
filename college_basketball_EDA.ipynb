```python
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/subwaymatch/my-project-name/raw/main/college-basketball-2021.csv')
```

```python
df.head(3)
```

```python
df_average_conference_seed = df.groupby('CONF', as_index=False).agg({
    'SEED': ['count', 'mean', 'max', 'min']
})

df_average_conference_seed.head(3)
```

```python
df_average_conference_seed.columns
```

```python
df.head(3)
```

```python
df_average_conference_seed.sort_values(('SEED', 'count'), inplace=True)
```

```python
df_average_conference_seed.columns = ['CONF', 'SEED_COUNT', 'SEED_MEAN', 'SEED_MAX', 'SEED_MIN']
df_average_conference_seed.head(3)
```
